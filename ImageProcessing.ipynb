```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
from glob import glob
import os

path_to_train = ''
glob_train_imgs = os.path.join(path_to_train, '*.tiff')


train_img_paths = glob(glob_train_imgs)

print(train_img_paths[:10])

```

```python
im = Image.open('WednesdayA-1_b0s0c0x45703-1600y6679-1200m350.tiff')
im.show()
imarray = np.array(im)
print (imarray.shape)
```

```python
im = Image.open(train_img_paths[0]) #first image outside of loop because you concatenate 
start = np.array(im)
plt.imshow(start/255)
plt.show()
for i in range(1,len(train_img_paths)):
    #print (train_img_paths[i])
    im = Image.open(train_img_paths[i])
    imarray = np.array(im)
    plt.imshow(imarray/255)
    plt.show()
    start = np.hstack((start, imarray))
print (start.shape)
```
